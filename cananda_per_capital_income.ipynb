```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# import csv file
df = pd.read_csv("C:\\Users\\NTS-PratikshaSarole\\Downloads\\canada_per_capita_income.csv")
df.head(4)
```

```python
# scatter plot of year 
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year, df['per capita income (US$)'],color='blue')
```

```python
#create linear regression
reg = linear_model.LinearRegression()
```

```python
#fit the model
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
# predicting the price for given area size
prediction =reg.predict([[2020]])
prediction
```

```python
# find coeff 
reg.coef_
```

```python
# find intercept
reg.intercept_
```

```python
# linear equation mx+c:(per capita income (US$)) = (reg.coef_  * df(['year'] + reg.intercept_)
# prediction for captial income in year 2020
prediction = 828.46507522 * 2020 + (-1632210.7578554575)
prediction
```

```python
# linear regression chart
plt.xlabel('Year',fontsize=20)
plt.ylabel('Per Capita Income (US$)',fontsize=20)
plt.scatter(df.year,df[['per capita income (US$)']],color='blue')
plt.plot(df.year,reg.predict(df[['year']]),color='red')
plt.show()
```
